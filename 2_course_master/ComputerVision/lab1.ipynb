## <p style="text-align:center">Загрузка изображений</p>

Для загрузки изображений используется функция cv2.imread().

Аргументы функции cv2.imread():

* filename - путь до исходного изображения (полный или относительный)
* flags - необязательные флаги, которые позволяют открыть изображение в нужном формате:
    - cv2.IMREAD_COLOR: загружает цветное изображение. Прозрачностью изображения можно пренебречь. Флаг по умолчанию. Можно заменить параметром: 1
    - cv2.IMREAD_GRAYSCALE: загружает изображение в режиме оттенков серого. Можно заменить параметром: 0
    - cv2.IMREAD_UNCHANGED: загружает изображение как таковое, включая альфа-канал. Можно заменить параметром: -1

Для отображения изображения в новом окне используется функция cv2.imshow().

Аргументы функции cv2.imshow():

* winname: имя окна в строковом формате.
* mat: объект изображения.

Для привязки клавиатуры используется функция cv2.waitKey(). Аргументом является время в миллисекундах. Если передается 0, то ожидание нажатия клавиши будет бесконечно долгим. Также может быть настроено обнаружение определенных нажатий клавиш.

Функция cv2.destroyAllWindows() уничтожает все созданные нами окна.

```python
# Подключение библиотек
import cv2


# Загрузка изображения в исходном виде
image = cv2.imread("./путь/к/изображению.расширение")
# Отображение результата в новом окне и ожидание нажатия кнопки пользователем для закрытия окна
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
