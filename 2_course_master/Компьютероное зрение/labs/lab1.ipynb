## Загрузка изображения

Для загрузки изображения используется функция imread(). Параметры функции перечислены ниже:
* filename: путь к файлу изображения в строковом формате;
* flags: дополнительные флаги для открытия изображения:
    * cv2.IMREAD_COLOR (1) - флаг открытия изображения в цветном формате (используется по умолчанию);
    * cv2.IMREAD_GRAYSCALE (0) - флаг открытия изображения в градациях серого;
    * cv2.IMREAD_UNCHANGED (-1) - флаг открытия изображения с включением альфа-канала.

Для отображения используется функция imshow(), которая включает в себя следующие параметры:
* winname: название окна, внутри которого откроется изображение, в строковом формате;
* mat: MatLike объект, который содержит изображение, которое нужно отобразить (все изображения в OpenCV являются MatLike).

Для взаимодействия с клавиатурой используется функция waitKey(), единственным параметром которой является время задержки для каждого нажатия.

Для закрытия всех окон используется функция destroyAllWindows(), у которой нет параметров.

Пример работы этих функций показан ниже:

```python
import cv2


# Название тестовой картинки
pic = '../resources/test_picture.jpg'

# Открытие изображени
image = cv2.imread(pic)
# Показ изображения в отдельном окне
cv2.imshow("Test picture", image)
# Ожидание нажатия клавиши для закрытия окна с изображением
cv2.waitKey(0)
cv2.destroyAllWindows()
```

На фрагменте кода ниже представлена оформленная функция для отображения изображений.

```python
from cv2.typing import MatLike


def viewImage(image: MatLike, name_of_window: str):
    '''Функция для открытия окна с указанным изображением и заголовком'''
    cv2.namedWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


viewImage(image, 'Test picture')
```

# Сохранение изображений
Для сохранения изображений используется функция imwrite(), у которой есть следующие аргументы:
* filename: путь до файла, в который необходимо сохранить изображение (файл создается или перезаписывается);
* img: MatLike объект с иображением, которое необходимо записать;
* params: дополнительные параметры для сохранения изображения (здесь и далее не используются).

Фрагмент кода, иллюстрирующий сохранение в файл приведен ниже:

```python
img = cv2.imread(pic, 0)
cv2.imshow('image', img)
k = cv2.waitKey(0)
if k == 27:         # Ожидание нажатия клавиши Esc для закрытия окна с изображением
    cv2.destroyAllWindows()
elif k == ord('s'): # Ожидание нажатия клавиши s для сохранения изображения в градациях серого
    cv2.imwrite('../resources/test_gray_picture.png', img)
    cv2.destroyAllWindows()
```

# Работа с видеопотоком
Для захвата видео необходимо создать объект VideoCapture, аргументами которого может быть:
* index: индекс устройства, либо имя видеофайла;
* apiPreference: здесь и далее не используется.

Для установки пользовательских настроек используется метод set(). Этот метод позволяет задать следующие настройки:
* CV_CAP_PROP_POS_MSEC: текущая позиция видеофайла в миллисекундах или временная метка захвата видео;
* CV_CAP_PROP_POS_FRAMES: 0-индекс кадра, который будет декодирован/захвачен следующим;
* CV_CAP_PROP_POS_AVI_RATIO: относительное положение видеофайла: 0 - начало фильма, 1 - Конец фильма;
* CV_CAP_PROP_FRAME_WIDTH: ширина кадров в видеопотоке;
* CV_CAP_PROP_FRAME_HEIGHT: высота кадров в видеопотоке;
* CV_CAP_PROP_FPS: частота кадров;
* CV_CAP_PROP_FOURCC: 4-символьный код кодека;
* CV_CAP_PROP_FRAME_COUNT: количество кадров в видеофайле;
* CV_CAP_PROP_FORMAT: формат объектов Mat, возвращаемых функцией retrieve();
* CV_CAP_PROP_MODE: Backend-специфичное значение, указывающее текущий режим захвата;
* CV_CAP_PROP_BRIGHTNESS: яркость изображения (только для камер);
* CV_CAP_PROP_CONTRAST: контрастность изображения (только для камер);
* CV_CAP_PROP_SATURATION: насыщенность изображения (только для камер);
* CV_CAP_PROP_HUE: оттенок изображения (только для камер);
* CV_CAP_PROP_GAIN: усиление изображения (только для камер);
* CV_CAP_PROP_EXPOSURE: экспозиция (только для камер);
* CV_CAP_PROP_CONVERT_RGB: булевы флаги, указывающие, является ли изображения должны быть преобразованы в RGB;
* CV_CAP_PROP_WHITE_BALANCE_U: значение U настройки баланса белого (Примечание: В настоящее время поддерживается только бэкендом DC1394 v 2.x);
* CV_CAP_PROP_WHITE_BALANCE_V: значение V настройки баланса белого (Примечание: В настоящее время поддерживается только бэкендом DC1394 v 2.x);
* CV_CAP_PROP_RECTIFICATION: флаг выпрямления для стереокамер (Примечание: В настоящее время поддерживается только бэкендом DC1394 v 2.x);
* CV_CAP_PROP_ISO_SPEED: скорость ISO камеры (Примечание: В настоящее время поддерживается только бэкендом DC1394 v 2.x);
* CV_CAP_PROP_BUFFER: количество кадров, хранящихся во внутренней буферной памяти (Примечание: В настоящее время поддерживается только бэкендом DC1394 v 2.x)/

Для чтения видеопотока кадр за кадром используется метод read(), единственным методом которого может быть только объект изображения.

Для того, чтобы корректно завершить работу с камерой нужно отпустить захват видеопотока и это делается с использвоанием метода release().

Пример, иллюстрирующий захват видеопотока приведен ниже:

```python
# Создание объекта видеопотока
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FPS, 24) # Частота кадров
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 600) # Ширина кадров в видеопотоке.
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480) # Высота кадров в видеопотоке.

while(True):
    # Захват изображения кадр за кадром
    ret, frame = cap.read()

    # Градации серого для каждого кадра
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Показ результата обработки в реальном времени на экран
    cv2.imshow('frame', gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Отпускаем захват видеопотока
cap.release()
cv2.destroyAllWindows()
```

# Запись видео с камеры
Для записи видео используется класс VideoWriter, в частности, метод fourcc(), который получает идентификатор кодека.

Аргументы класса:
* filename: путь к файлу для записи видеопотока;
* fourcc: идентификатор кодека, которым кодируется видео;
* fps: частота кадров для записи;
* frameSize: размер фрейма для записи (разрешение);
* isColor: флаг для записи цветного/черно-белого видеопотока.

Для записи кадра используется метод write(), аргументом для которого является, собственно, кадр.

Несколько фрагментов кода, иллюстрирующих разные настройки камеры приведены далее:

```python
cap = cv2.VideoCapture(0)
# Устанавливаем кодек для кодирования видео
codec = cv2.VideoWriter.fourcc(*'XVID')
# Устанавливаем параметры для записываемого видео
out = cv2.VideoWriter('../resources/captured.avi', codec, 25.0, (640, 480))
while(cap.isOpened()):
    ret, frame = cap.read()
    if cv2.waitKey(1) & 0xFF == ord('q') or ret == False:
        break
    cv2.imshow('frame', frame)
out.write(frame)
out.release()
cap.release()
cv2.destroyAllWindows()
codec = cv2.VideoWriter.fourcc(*'XVID')
```

```python
cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter.fourcc(*'XVID')
out = cv2.VideoWriter('../resources/output.avi', fourcc, 20.0, (640, 480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        # Переворачиваем кажддый кадр и записываем его
        frame = cv2.flip(frame,0)
        out.write(frame)

        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Отпускаем захват в конце работы программы
cap.release()
out.release()
cv2.destroyAllWindows()
```

# Воспроизведение видео из файла
В данной теме можно рассмотреть использование метода cvtColor().

cvtColor() — метод, который преобразует изображение из одного цветового пространства в другое. Аргументы метода:
* src: объект изображения для преобразования;
* code: код для преобразования между цветовыми пространствами;
* dst: выходное изображение того же размера и глубины, что и исходное (является также возвращаемым);
* dstCn: количество каналов в целевом изображении (здесь и далее не используется);
* hint: флаг модификации реализации AlgorithmHint;

Данный метод используется для преобразования конкретного кадра.

Пример реализации воспроизведения видеофрагмента показан ниже:

```python
# В этот раз в качесве параметра передвем одно из записанных видео
cap = cv2.VideoCapture('../resources/output.avi')

while(cap.isOpened()):
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('frame', gray)
    # Изменяем время для корректной скорости видео
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Функции рисования в OpenCV
Для рисования в OpenCV применяются следующие методы line(), circle(), rectangle(), ellipse(), putText() и некоторые другие.

Для всех этих функций есть общий набор арггументов, среди которых:
* img: исходное изображение;
* color: цвет фигуры; для BGR передается кортеж, а для оттенков серого скалярное значение;
* thickness: толщина линии, круга и т. д.; если для замкнутых фигур, таких как круги, передается значение -1, то оно заполняет фигуру; толщина по умолчанию = 1;
* lineType: тип линии, по умолчанию это 8-connected.

Также, в зависимости от фигуры могут подаваться кортежи точек или координат центра и радиуса.

Пример рисования объектов представлен в коде ниже:

```python
import numpy as np


# Создание пустого черного квадрата
img = np.zeros((512, 512, 3), np.uint8)

# Рисуем диагональную линию с толщиной 5 пикселей
img = cv2.line(
    img,
    (0, 0),
    (511, 511),
    (255, 0, 0),
    5
)
# рисуем зеленый прямоугольник в правом верхнем углу толщиной 3 пикселя
img = cv2.rectangle(
    img,
    (384, 0),
    (510, 128),
    (0, 255, 0),
    3
)
# Рисуем красный круг внутри прямоугольника
img = cv2.circle(
    img,
    (447, 63),
    63,
    (0, 0, 255),
    -1
)
# Рисуем полуэлипс в центре
img = cv2.ellipse(
    img,
    (256, 256),
    (100, 50),
    0,
    0,
    180,
    255,
    -1
)
# Рисуем незамкнутый многоугольник желтого цвета
pts = np.array(
    [[10, 5], [20, 30], [70, 20], [50, 10]],
    np.int32
)
pts = pts.reshape((-1, 1, 2))
img = cv2.polylines(img, [pts], True, (0, 255, 255))
# Пишем текст снизу изображения
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(
    img,
    'OpenCV',
    (10, 500),
    font,
    4,
    (255, 255, 255),
    2,
    cv2.LINE_AA
)

cv2.imshow('lime', img)

# Отрисовываем окошко
if cv2.waitKey(10000) & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```
