```python
import cv2


# Название тестовой картинки
pic = 'test_picture.jpg'

# Открытие изображени
image = cv2.imread(pic)
# Показ изображения в отдельном окне
cv2.imshow("Test picture", image)
# Ожидание нажатия клавиши для закрытия окна с изображением
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from cv2.typing import MatLike


def viewImage(image: MatLike, name_of_window: str):
    '''Функция для открытия окна с указанным изображением и заголовком'''
    cv2.namedWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


viewImage(image, 'Test picture')
```

```python
img = cv2.imread(pic, 0)
cv2.imshow('image', img)
k = cv2.waitKey(0)
if k == 27:         # Ожидание нажатия клавиши Esc для закрытия окна с изображением
    cv2.destroyAllWindows()
elif k == ord('s'): # Ожидание нажатия клавиши s для сохранения изображения в градациях серого
    cv2.imwrite('test_gray_picture.png',img)
    cv2.destroyAllWindows()
```

```python
# Создание объекта видеопотока
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FPS, 24) # Частота кадров
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 600) # Ширина кадров в видеопотоке.
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480) # Высота кадров в видеопотоке.

while(True):
    # Захват изображения кадр за кадром
    ret, frame = cap.read()

    # Градации серого для каждого кадра
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Показ результата обработки в реальном времени на экран
    cv2.imshow('frame', gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Отпускаем захват видеопотока
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
# Устанавливаем кодек для кодирования видео
codec = cv2.VideoWriter_fourcc(*'XVID')
# Устанавливаем параметры для записываемого видео
out = cv2.VideoWriter('captured.avi',codec, 25.0, (640,480))
while(cap.isOpened()):
    ret, frame = cap.read()
    if cv2.waitKey(1) & 0xFF == ord('q') or ret == False:
        break
    cv2.imshow('frame', frame)
out.write(frame)
out.release()
cap.release()
cv2.destroyAllWindows()
codec = cv2.VideoWriter_fourcc(*'XVID')
```

```python
cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        # Переворачиваем кажддый кадр и записываем его
        frame = cv2.flip(frame,0)
        out.write(frame)

        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Отпускаем захват в конце работы программы
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
# В этот раз в качесве параметра передвем одно из записанных видео
cap = cv2.VideoCapture('output.avi')

while(cap.isOpened()):
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('frame',gray)
    # Изменяем время для корректной скорости видео
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np


# Создание пустого черного квадрата
img = np.zeros((512, 512, 3), np.uint8)

# Рисуем диагональную линию с толщиной 5 пикселей
img = cv2.line(
    img,
    (0, 0),
    (511, 511),
    (255, 0, 0),
    5
)
# рисуем зеленый прямоугольник в правом верхнем углу толщиной 3 пикселя
img = cv2.rectangle(
    img,
    (384, 0),
    (510, 128),
    (0, 255, 0),
    3
)
# Рисуем красный круг внутри прямоугольника
img = cv2.circle(
    img,
    (447, 63),
    63,
    (0, 0, 255),
    -1
)
# Рисуем полуэлипс в центре
img = cv2.ellipse(
    img,
    (256, 256),
    (100, 50),
    0,
    0,
    180,
    255,
    -1
)
# Рисуем незамкнутый многоугольник желтого цвета
pts = np.array(
    [[10, 5], [20, 30], [70, 20], [50, 10]],
    np.int32
)
pts = pts.reshape((-1, 1, 2))
img = cv2.polylines(img, [pts], True, (0, 255, 255))
# Пишем текст снизу изображения
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(
    img,
    'OpenCV',
    (10, 500),
    font,
    4,
    (255, 255, 255),
    2,
    cv2.LINE_AA
)

cv2.imshow('lime', img)

# Отрисовываем окошко
if cv2.waitKey(10000) & 0xFF == ord('q'):
    cv2.destroyAllWindows()
```
